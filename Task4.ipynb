```python
# adding safety filter to block harmful/dangerous medical queries

def safety_filter(user_input):
    dangerous_phrases = [
        "suicide",
        "kill myself",
        "overdose",
        "self harm",
        "how much medicine should i take",
        "replace doctor",
        "skip doctor",
        "what medicine should i take",
        "suggest medicine"
    ]

    for phrase in dangerous_phrases:
        if phrase in user_input.lower():
            return False
    return True
```

```python
# system prompt to control behavior and tone of the chatbot

SYSTEM_PROMPT = """
You are a helpful and friendly medical assistant.
Provide general health information only.
Do NOT diagnose diseases.
Do NOT prescribe medicine.
Do NOT give exact dosages.
Encourage consulting a healthcare professional when necessary.
Keep answers simple and easy to understand.
"""
```
